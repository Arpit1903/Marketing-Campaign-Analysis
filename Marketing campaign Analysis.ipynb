```python
df_campaign_perf["ROI_Category"] = np.where(df_campaign_perf['ROI'] > 0.5, "High ROI", np.where(df_campaign_perf['ROI'] > 0.2, "Medium ROI", "Low ROI"))
```

```python
print(df_campaign_perf["ROI_Category"])
```

```python
df_campaign_perf['Performance'] = np.where(
    df_campaign_perf['Revenue'] > df_campaign_perf['Revenue'].median(),
    'Above Average',
    'Below Average'
)
```

```python
print(df_campaign_perf['Performance'])
```

```python
avg_roi = np.mean(df_campaign_perf['ROI'])
median_roi = np.median(df_campaign_perf['ROI'])
std_roi = np.std(df_campaign_perf['ROI'])
percentile_90 = np.percentile(df_campaign_perf['ROI'], 90)
```

```python
print("Average ROI is: ", avg_roi)
print("Median ROI is: ", median_roi)
print("Standard Deviation ROI is: ", std_roi)
print("90th Percentile ROI is: ", percentile_90)
```
